```python
#1.  Find the average ticket price per airline.
import pandasql as ps

query = """
SELECT airline, AVG(price) as avg_price
FROM df
WHERE source_city = 'Delhi'
GROUP BY airline
ORDER BY avg_price DESC
LIMIT 5;
"""

result = ps.sqldf(query, locals())
print(result)
```

```python
#2. List the top 5 most expensive routes (source to destination).
import pandasql as ps

query = """
SELECT source_city, destination_city, AVG(Price) AS AvgPrice
FROM df
GROUP BY source_city, destination_city
ORDER BY AvgPrice DESC
LIMIT 5;
"""

result = ps.sqldf(query, locals())
print(result)
```

```python
#3. Which airline offers the lowest average ticket price for Business class?
import pandasql as ps

query = """
SELECT airline, AVG(Price) AS AvgBusinessPrice
FROM df
WHERE Class = 'Business'
GROUP BY airline
ORDER BY AvgBusinessPrice ASC
LIMIT 1;
"""

result = ps.sqldf(query, locals())
print(result)
```

```python
#4. Find routes with more than 50 flights and their average ticket prices.
import pandasql as ps

query = """
SELECT source_city, destination_city, 
    COUNT(*) AS FlightCount, 
    AVG(Price) AS AvgPrice
FROM df
GROUP BY source_city, destination_city
HAVING COUNT(*) > 50;
"""

result = ps.sqldf(query, locals())
print(result)
```

```python
#5. Compare average price difference between Economy and Business class per airline.
import pandasql as ps

query = """
SELECT 
    e.airline,
    AVG(CASE WHEN e.Class = 'Economy' THEN e.Price END) AS AvgEconomyPrice,
    AVG(CASE WHEN e.Class = 'Business' THEN e.Price END) AS AvgBusinessPrice,
    (AVG(CASE WHEN e.Class = 'Business' THEN e.Price END) -
     AVG(CASE WHEN e.Class = 'Economy' THEN e.Price END)) AS PriceDifference
FROM df e
GROUP BY e.airline;
"""

result = ps.sqldf(query, locals())
print(result)
```

```python
#6. Which cities have the most incoming flights? 
import pandasql as ps

query = """
SELECT 
    destination_city, 
    COUNT(*) AS FlightCount
FROM df
GROUP BY destination_city
ORDER BY FlightCount DESC
LIMIT 10;
"""

result = ps.sqldf(query, locals())
print(result)
```

```python
7. Which cities have the most outgoing flights?
import pandasql as ps

query = """
SELECT source_city, 
    COUNT(*) AS FlightCount
FROM df
GROUP BY source_city
ORDER BY FlightCount DESC
LIMIT 10;
"""

result = ps.sqldf(query, locals())
print(result)
```

```python
#8. Find the busiest route (most number of flights)
import pandasql as ps

query = """
SELECT source_city, destination_city, 
    COUNT(*) AS FlightCount
FROM df
GROUP BY source_city, destination_city
ORDER BY FlightCount DESC
LIMIT 1;
"""

result = ps.sqldf(query, locals())
print(result)
```

```python
#9. List the top 3 airlines with cheapest average prices for each route.
import pandasql as ps 

query = """
SELECT source_city, destination_city, airline, AvgPrice
FROM (
    SELECT 
        source_city,
        destination_city,
        airline,
        AVG(Price) AS AvgPrice,
        ROW_NUMBER() OVER (
            PARTITION BY source_city, destination_city 
            ORDER BY AVG(Price) ASC
        ) AS rn
    FROM df
    GROUP BY source_city, destination_city, airline
) t
WHERE rn <= 3
ORDER BY source_city, destination_city, AvgPrice;
"""

result = ps.sqldf(query, locals())
print(result)
```

```python
#10. Find flights with duration more than 5 hours but priced below average.
import pandasql as ps

query = """
SELECT *
FROM df
WHERE Duration > 5
  AND Price < (SELECT AVG(Price) FROM df);
  """

result = ps.sqldf(query, locals())
print(result)
```

```python
#11. Which flight class shows the highest average price variation (std deviation)?
import pandasql as ps

query = """
SELECT 
    class, 
    AVG(Price) AS avgPrice
FROM data
GROUP BY class
ORDER BY avgprice DESC;
"""

result = ps.sqldf(query, locals())
print(result)
```

```python
#12. Find how ticket prices vary with number of days left before departure.
import pandasql as ps

query = """
SELECT days_Left, AVG(Price) AS AvgPrice
FROM df
GROUP BY days_Left
ORDER BY days_Left DESC;
"""

result = ps.sqldf(query, locals())
print(result)
```

```python
#13. Which time of day (Morning, Evening, etc.) has the highest average ticket price?
import pandasql as ps

query = """
SELECT departure_time,AVG(Price) AS AvgPrice
FROM data
GROUP BY departure_time
ORDER BY AvgPrice DESC
LIMIT 5;
"""

result = ps.sqldf(query, locals())
print(result)
```

```python
#14. Find Price trends by airline and travel class
import pandasql as ps

query = """
SELECT 
    airline,
    class,
    AVG(price) AS avg_price,
    MIN(price) AS min_price,
    MAX(price) AS max_price
FROM df
GROUP BY airline, class
ORDER BY airline, class;
"""

result = ps.sqldf(query, locals())
print(result)
```

```python
#15. Compare the airlines average prices for non-stop vs. 1-stop vs. 2+ stops flights.
import pandasql as ps

query = """
SELECT 
    airline,
    CASE 
        WHEN Stops = 0 THEN 'Non-stop'
        WHEN Stops = 1 THEN '1-stop'
        ELSE '2+ stops'
    END AS StopsCategory,
    AVG(Price) AS AvgPrice
FROM df
GROUP BY airline, StopsCategory
ORDER BY airline, StopsCategory;
"""

result = ps.sqldf(query, locals())
print(result)
```
